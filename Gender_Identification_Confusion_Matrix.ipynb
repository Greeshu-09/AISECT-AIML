```python
#Importing pandas for data analaysis
import pandas as pd
```

```python

#loading data and assigning it a variable name"dataset"
dataset=pd.read_csv("/content/gender_classification_v7.csv")
```

```python
dataset
```

```python
dataset
```

```python

#splitting the value for independent and independent
x=dataset[["long_hair",	"forehead_width_cm",	"forehead_height_cm",	"nose_wide",	"nose_long",	"lips_thin",	"distance_nose_to_lip_long"]]
y=dataset[["gender"]]
```

```python
#splitting the value 80% independent and 20% dependent
from sklearn.model_selection import train_test_split
```

```python

X_train,X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
#importing the algorithm-here we choose decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
cls=DecisionTreeClassifier()
cls.fit(X_train,y_train)
```

```python
y_pred=cls.predict(X_test)
```

```python
##importing performance metrics since we use classification algorithm
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
```

```python
precision = precision_score(y_test, y_pred,average=None,labels=["Male","Female"])
recall = recall_score(y_test, y_pred,average=None,labels=["Male","Female"])
```
