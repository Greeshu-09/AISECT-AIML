```python
#Importing pandas for data analaysis
import pandas as pd
```

```python
#loading data and assigning it a variable name"dataset"
dataset=pd.read_csv("/content/data.csv")
```

```python
dataset
```

```python
dataset.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
dataset[["city"]].value_counts()
```

```python
le = LabelEncoder()
dataset["city_encoded"] = le.fit_transform(dataset["city"])
```

```python
dataset_model = dataset.drop(columns=[ "date", "street", "statezip", "country","city"])
```

```python
dataset_model
```

```python
x=dataset[["bedrooms","bathrooms",	"sqft_living",	"sqft_lot",	"floors",	"waterfront",	"view",	"condition",	"sqft_above",	"sqft_basement",	"yr_built",	"yr_renovated",	"city_encoded"]]
y=dataset[["price"]]
```

```python
#importing train test split from scikit
from sklearn.model_selection import train_test_split
```

```python
#the dataset is splitted (80% for training and 20% for testing)
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
from sklearn.tree import DecisionTreeRegressor

```

```python
reg=DecisionTreeRegressor()
reg.fit(X_train,y_train)
```

```python
y_pred=reg.predict(X_test)
```

```python
#to check the accuracy
from sklearn.metrics import r2_score
```

```python
#to find the r2score
r2=r2_score(y_test,y_pred)
```

```python
r2
```

```python
#importing pickle to save the file
import pickle
```

```python
#saving the file under sav extension
filename="DecisionTreeRegressor.sav"
```

```python
pickle.dump(reg,open(filename,"wb"))
```
