```python
#Importing pandas for data analaysis
import pandas as pd
```

```python
#loading data and assigning it a variable name"dataset"
dataset=pd.read_csv("/content/MBA.csv")
```

```python
dataset
```

```python
#checking if there is any null values in dataset
dataset.isnull().sum()
```

```python
#This step is to check the type of encoding we need to use
dataset[["major"]].value_counts()
```

```python
#since the labels are more than 2 we use label encoder
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
#Encoding 2 columns
dataset["major"]=le.fit_transform(dataset["major"])
dataset["work_industry"]=le.fit_transform(dataset["work_industry"])
```

```python
#dropping the unnecessary columns
dataset=dataset.drop(columns=["race"])
```

```python
dataset=dataset.drop(columns=["gender","international"])
```

```python
dataset
```

```python
#splitting the value for independent and independent
x=dataset[["gpa","major",	"gmat",	"work_exp",	"work_industry"]]
y=dataset[["admission"]]
```

```python
#splitting the value 80% independent and 20% dependent
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
#importing the algorithm-here we choose decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
cls=DecisionTreeClassifier()
cls.fit(X_train,y_train)
```

```python
y_pred=cls.predict(X_test)
```

```python
#importing pickle to save the file
import pickle
```

```python
#saving the file under sav extension
filename="DecisionTreeclassifier.sav"
```

```python

pickle.dump(cls,open(filename,"wb"))
```

```python
#reading the code which we saved
load_model=pickle.load(open("/content/DecisionTreeclassifier.sav","rb"))
```
