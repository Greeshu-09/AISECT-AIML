```python
#Importing pandas for data analaysis
import pandas as pd
```

```python
#loading data and assigning it a variable name"dataset"
dataset=pd.read_csv("/content/college.csv")
```

```python
dataset
```

```python
#Checking the label under this category to choose the type of encoding
dataset[["Encourage"]].value_counts()
```

```python
#Choosing onehot coding as there are only two labels
from sklearn.preprocessing import OneHotEncoder
one=OneHotEncoder(handle_unknown='ignore')
dataset=pd.get_dummies(dataset,columns=['Gender','Encourage'],drop_first=True,dtype=int)
```

```python
dataset
```

```python
#splitting the value for independent and independent
x=dataset[["Parent_income",	"IQ",	"Gender_male",	"Encourage_not encourage"]]
y=dataset[["Plan"]]
```

```python
#splitting the value 80% independent and 20% dependent
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test, y_train, y_test = train_test_split(x, y, test_size=0.20, random_state=42)
```

```python
#importing the algorithm-here we choose decision tree
from sklearn.tree import DecisionTreeClassifier
```

```python
cls=DecisionTreeClassifier()
cls.fit(X_train,y_train)
```

```python
y_pred=cls.predict(X_test)
```

```python
#importing performance metrics since we use classification algorithm
from sklearn.metrics import precision_score
from sklearn.metrics import recall_score
from sklearn.metrics import f1_score
```

```python
precision = precision_score(y_test, y_pred,average=None)
recall = recall_score(y_test, y_pred,average=None)
f1=f1_score(y_test, y_pred,average=None)
```
